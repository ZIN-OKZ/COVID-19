```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
import datetime as dt
covid_data = pd.read_csv('https://stopcovid19.metro.tokyo.lg.jp/data/130001_tokyo_covid19_patients.csv')
```

```python
print("最終更新日："+covid_data["公表_年月日"].max())
```

```python
covid_data["患者_性別"] = covid_data["患者_性別"].replace('男', '男性')
covid_data["Date"] = pd.to_datetime(covid_data["公表_年月日"])
covid_data["Month"] = covid_data["Date"].dt.strftime("%Y-%m")
covid_data["MMDD"] = covid_data["Date"].dt.strftime("%m%d")
```

```python
today = covid_data[covid_data["Date"] >= covid_data["公表_年月日"].max()].count()["No"].astype(str)
print(covid_data["公表_年月日"].max() + "の感染者数（東京都）：" + today + "名")
```

```python
print("これまでの累計感染者数（東京都）：" + covid_data.count()["No"].astype(str) + "名")
```

```python
covid_data.groupby("Month").count()[["No"]]
```

```python
covid_data.groupby("MMDD").count()[["No"]].plot(kind="bar", grid=True, rot=90, fontsize=6, legend=False)
plt.xlabel("Day" ,size=12, fontname="Meiryo")
plt.title("東京都感染者数（日別）", fontname="Meiryo")
```

```python
Daily = covid_data.groupby("MMDD").count()[["No"]]
Daily.cumsum().plot(kind="bar", grid=True, rot=90, fontsize=6, legend=False)
plt.xlabel("Day" ,size=12)
plt.title("東京都感染者数（累計）", fontname="Meiryo")
```
